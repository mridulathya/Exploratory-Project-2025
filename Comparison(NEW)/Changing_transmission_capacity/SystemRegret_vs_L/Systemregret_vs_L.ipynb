```python
import matplotlib.pyplot as plt
import numpy as np


# X-tick labels in LaTeX
num_files = [r"$F'$", r"$\frac{F'}{3}$", r"$\frac{F'}{4}$", r"$\frac{F'}{5}$"]                                                                                                                                                                                      
x = np.arange(len(num_files))

# Fonts and layout
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 14,
    "axes.labelsize": 14,
    "legend.fontsize": 12,
    "xtick.labelsize": 13,
    "ytick.labelsize": 13,
})

fig, ax = plt.subplots(figsize=(7, 4), dpi=300)

# Distinct line styles with markers
ax.plot(x, [qier[0],qier[2],qier[3],qier[4]], label='H-DDQN-QiER', color='black', linestyle='-', marker='o', markersize=6)
ax.plot(x, [per[0],per[2],per[3],per[4],],  label='H-DDQN-PER',  color='red', linestyle='--', marker='s', markersize=6)
ax.plot(x, [lru[0], lru[2],lru[3],lru[4]],  label='H-DDQN-LRU',  color='blue', linestyle='-.', marker='^', markersize=6)
ax.plot(x, [fifo[0], fifo[2], fifo[3], fifo[4]], label='H-DDQN-FIFO', color='green', linestyle=':', marker='D', markersize=6)

# Axes and ticks
ax.set_xlabel('Transmission capacity')
ax.set_ylabel('Avg. System regret')
ax.set_xticks(x)
ax.set_xticklabels(num_files)
ax.set_ylim(0.2, 0.6)

# Grid and legend
ax.grid(True, which='major', axis='y', linestyle='--', alpha=0.7)
ax.legend(loc='best', frameon=1)

fig.tight_layout()
plt.savefig("avg_Systemregret_vs_transmission_capacity_linegraph.png", dpi=300)
plt.show()
```
