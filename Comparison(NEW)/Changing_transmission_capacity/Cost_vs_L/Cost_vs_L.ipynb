```python
import matplotlib.pyplot as plt
import numpy as np

# Fonts and layout
plt.rcParams.update({
    "font.family": "serif",
    "font.size":14,
    "axes.labelsize":14,
    "legend.fontsize": 14,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
})

x = np.arange(len(num_files))
width = 0.15  # Reduced width to fit 4 bars per group

fig, ax = plt.subplots(figsize=(7, 4), dpi=300)

# Styled bars
bars1 = ax.bar(x - 1.5*width, qier, width, label='H-DDQN-QiER', color='gold', hatch='..', edgecolor='black')
bars2 = ax.bar(x + 0.5*width, per,  width, label='H-DDQN-PER',  color='magenta', hatch='\\\\', edgecolor='black')
bars3 = ax.bar(x + 1.5*width, lru,  width, label='H-DDQN-LRU',  color='limegreen', hatch='//', edgecolor='black')
bars4 = ax.bar(x -0.5*width, fifo, width, label='H-DDQN-FIFO', color='steelblue', hatch='O', edgecolor='black')

# Axes labels and ticks
ax.set_xlabel('Transmission capacity')
ax.set_ylabel('Avg. CSP cost')
ax.set_xticks(x)
# ax.set_ylim(0.6,1.0)
ax.set_xticklabels(num_files)

# Grid and legend
ax.grid(axis='y', linestyle='--', alpha=0.3)
ax.legend( loc='best')

fig.tight_layout()
plt.show()
plt.savefig("avg_COST_vs_transmission_capacity.png", dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fonts and layout
plt.rcParams.update({
    "font.family": "serif",
    "font.size":14,
    "axes.labelsize":14,
    "legend.fontsize":14,
    "xtick.labelsize":14,
    "ytick.labelsize":14,
})

fig, ax = plt.subplots(figsize=(7, 4), dpi=300)

# Styled lines
ax.plot(num_files, qier, marker='*', color='red', label='H-DDQN-QiER', linestyle='-', linewidth=2, markersize=10)
ax.plot(num_files, per,  marker='s', color='blue', label='H-DDQN-PER', linestyle='--', linewidth=2, markersize=6)
ax.plot(num_files, lru,  marker='^', color='limegreen', label='H-DDQN-LRU', linestyle='-.', linewidth=2, markersize=6)
ax.plot(num_files, fifo, marker='D', color='magenta', label='H-DDQN-FIFO', linestyle=':', linewidth=2, markersize=6)

# Axis labels and ticks
ax.set_xlabel('Transmission capacity')
ax.set_ylabel('Avg. CSP cost')
ax.set_xticks(num_files)
# ax.set_ylim(0.1, 0.8)

# Grid and legend
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend(frameon=True, loc='best')

fig.tight_layout()
plt.show()
plt.savefig("avg_chr_vs_transmission_capacity_linegraph.png", dpi=300)
```
