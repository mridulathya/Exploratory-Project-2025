```python
import matplotlib.pyplot as plt
import numpy as np

# Fonts and layout
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 14,
    "axes.labelsize": 14,
    "legend.fontsize": 14,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
})
fig, ax = plt.subplots(figsize=(7, 4), dpi=300)
x = [r"$1000$", r"$1010$", r"$1020$", r"$1030$"]   
# Styled lines
ax.plot(x, [qier[0],qier[1],qier[2],qier[3]], marker='o', color='#ffb000', label='H-DDQN-QiER', linestyle='-', linewidth=2, markersize=6)      # orange-red
ax.plot(x, [per[0],per[1],per[2],per[3]],  marker='s', color='#dc267f', label='H-DDQN-PER', linestyle='--', linewidth=2, markersize=6)     # purple
ax.plot(x, [lru[0],lru[1],lru[2],lru[3]],  marker='^', color='#648fff', label='H-DDQN-LRU', linestyle='-.', linewidth=2, markersize=6)     # green
ax.plot(x, [fifo[0],fifo[1],fifo[2],fifo[3]], marker='D', color='#00b386', label='H-DDQN-FIFO', linestyle=':', linewidth=2, markersize=6)     # blue

# Axis labels and ticks
ax.set_xlabel('No. of users')
ax.set_ylabel('Avg. Cache hit ratio')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.set_ylim(0.6, 1.0)

# Grid and legend
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend(frameon=True, loc='best', fontsize=12)

fig.tight_layout()
plt.show()
plt.savefig("avg_CHR_vs_uavs_linegraph.png", dpi=400)
```
